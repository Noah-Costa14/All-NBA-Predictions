```python
# THIS CODE BLOCK FIXES PLAYER NAMES IN DATAFRAMES

def filterNames(dataframe):
    playerArray = dataframe["Player"].array
    filteredPlyrArray = []
    for i in playerArray: 
        splitted = i.split("\\")
        filteredPlyrArray.append(splitted[0])
    NameSeries = pd.Series(filteredPlyrArray)
    dataframe["Player"] = NameSeries
    return dataframe

stats10 = filterNames(stats10)
stats11 = filterNames(stats11)
stats12 = filterNames(stats12)
stats13 = filterNames(stats13)
stats14 = filterNames(stats14)
stats15 = filterNames(stats15)
stats16 = filterNames(stats16)
stats17 = filterNames(stats17)
stats18 = filterNames(stats18)
stats16.head(10)
```

```python
# THIS CODE BLOCK FILTERS OUT UNWANTED PLAYERS FROM THE DATAFRAMES FOR VARIOUS REASONS

def filter_and_byPosition(dataframe):

    # Filters players out by their minutes played
    dataframe_minutes = dataframe[dataframe["MP"] > 24] # Would ideally want "GS" >= .8*G also, but cant figure it out unless I just filtered twice
    dataframe_filt = dataframe_minutes[dataframe_minutes["G"] >= 41]

    # Get the players who've been traded
    dataframe_tradedPlayers = dataframe_filt[dataframe_filt["Tm"] == "TOT"]["Player"].to_numpy()

    # Delete duplicate, non-TOT names from the df
    dataframe_drop_mask = (dataframe_filt["Player"].isin(dataframe_tradedPlayers)) & (dataframe_filt["Tm"] != "TOT")
    dataframe_filt = dataframe_filt.drop(dataframe_filt[dataframe_drop_mask].index, axis=0)

    # To check if deleting worked, uncomment this line. Should all be 1
    # print(dataframe_filt.groupby("Player")["Player"].count().sort_values(ascending=False))

    # Get the Centers, Guards, and Forwards
    dataframe_centers = dataframe_filt[dataframe_filt["Pos"] == "C"]
    dataframe_forwards = dataframe_filt[(dataframe_filt["Pos"] == "SF") | (dataframe_filt["Pos"] == "PF")]
    dataframe_guards = dataframe_filt[(dataframe_filt["Pos"] == "PG") | (dataframe_filt["Pos"] == "SG")]

    return dataframe_centers, dataframe_forwards, dataframe_guards

stats10_centers, stats10_forwards, stats10_guards = filter_and_byPosition(stats10)
stats11_centers, stats11_forwards, stats11_guards = filter_and_byPosition(stats11)
stats12_centers, stats12_forwards, stats12_guards = filter_and_byPosition(stats12)
stats13_centers, stats13_forwards, stats13_guards = filter_and_byPosition(stats13)
stats14_centers, stats14_forwards, stats14_guards = filter_and_byPosition(stats14)
stats15_centers, stats15_forwards, stats15_guards = filter_and_byPosition(stats15)
stats16_centers, stats16_forwards, stats16_guards = filter_and_byPosition(stats16)
stats17_centers, stats17_forwards, stats17_guards = filter_and_byPosition(stats17)
stats18_centers, stats18_forwards, stats18_guards = filter_and_byPosition(stats18)
stats21_centers, stats21_forwards, stats21_guards = filter_and_byPosition(stats21)
```

```python
# Ceates a dictionary by position that matches players with the number of times their name shows up

centersDict = {}
forwardsDict = {}
guardsDict = {}

for i in predicted_centers:
    if i not in centersDict.keys():
        centersDict[i] = 1
    else:
        centersDict[i] = centersDict.get(i) + 1

for j in predicted_forwards:
    if j not in forwardsDict.keys():
        forwardsDict[j] = 1
    else:
        forwardsDict[j] = forwardsDict.get(j) + 1

for k in predicted_guards:
    if k not in guardsDict.keys():
        guardsDict[k] = 1
    else:
        guardsDict[k] = guardsDict.get(k) + 1

# Sort players by their number of occurrences

best_centers = [k for k, v in sorted(centersDict.items(), key=lambda item: item[1])]
best_centers = best_centers[::-1][:3]
print(f"Predicted All-NBA Centers: {best_centers}")

best_forwards = [k for k, v in sorted(forwardsDict.items(), key=lambda item: item[1])]
best_forwards = best_forwards[::-1][:6]
print(f"Predicted All-NBA Forwards: {best_forwards}")

best_guards = [k for k, v in sorted(guardsDict.items(), key=lambda item: item[1])]
best_guards = best_guards[::-1][:6]
print(f"Predicted All-NBA Guards: {best_guards}")
```
