```python
# THIS CODE BLOCK FIXES PLAYER NAMES IN DATAFRAMES

def filterNames(dataframe):
    playerArray = dataframe["Player"].array
    filteredPlyrArray = []
    for i in playerArray: 
        splitted = i.split("\\")
        filteredPlyrArray.append(splitted[0])
    NameSeries = pd.Series(filteredPlyrArray)
    dataframe["Player"] = NameSeries
    return dataframe

stats10 = filterNames(stats10)
stats11 = filterNames(stats11)
stats12 = filterNames(stats12)
stats13 = filterNames(stats13)
stats14 = filterNames(stats14)
stats15 = filterNames(stats15)
stats16 = filterNames(stats16)
stats17 = filterNames(stats17)
stats18 = filterNames(stats18)


def encode(string):
    return unidecode.unidecode(string)

stats10["Player"] = stats10["Player"].apply(encode)
stats11["Player"] = stats11["Player"].apply(encode)
stats12["Player"] = stats12["Player"].apply(encode)
stats13["Player"] = stats13["Player"].apply(encode)
stats14["Player"] = stats14["Player"].apply(encode)
stats15["Player"] = stats15["Player"].apply(encode)
stats16["Player"] = stats16["Player"].apply(encode)
stats17["Player"] = stats17["Player"].apply(encode)
stats18["Player"] = stats18["Player"].apply(encode)

#stats21.head(180)
```

```python
# THIS CODE BLOCK FILTERS OUT UNWANTED PLAYERS FROM THE DATAFRAMES FOR VARIOUS REASONS

def filter_and_byPosition(dataframe):

    # Filters players out by their minutes played
    dataframe_minutes = dataframe[dataframe["MP"] > 24] # Would ideally want "GS" >= .8*G also, but cant figure it out unless I just filtered twice
    dataframe_filt = dataframe_minutes[dataframe_minutes["G"] >= 41]

    # Get the players who've been traded
    dataframe_tradedPlayers = dataframe_filt[dataframe_filt["Tm"] == "TOT"]["Player"].to_numpy()

    # Delete duplicate, non-TOT names from the df
    dataframe_drop_mask = (dataframe_filt["Player"].isin(dataframe_tradedPlayers)) & (dataframe_filt["Tm"] != "TOT")
    dataframe_filt = dataframe_filt.drop(dataframe_filt[dataframe_drop_mask].index, axis=0)

    # To check if deleting worked, uncomment this line. Should all be 1
    # print(dataframe_filt.groupby("Player")["Player"].count().sort_values(ascending=False))

    # Get the Centers, Guards, and Forwards
    dataframe_centers = dataframe_filt[dataframe_filt["Pos"] == "C"]
    dataframe_forwards = dataframe_filt[(dataframe_filt["Pos"] == "SF") | (dataframe_filt["Pos"] == "PF")]
    dataframe_guards = dataframe_filt[(dataframe_filt["Pos"] == "PG") | (dataframe_filt["Pos"] == "SG")]

    return dataframe_centers, dataframe_forwards, dataframe_guards

"""
stats10_centers, stats10_forwards, stats10_guards = filter_and_byPosition(result10)
stats11_centers, stats11_forwards, stats11_guards = filter_and_byPosition(result11)
stats12_centers, stats12_forwards, stats12_guards = filter_and_byPosition(result12)
stats13_centers, stats13_forwards, stats13_guards = filter_and_byPosition(result13)
stats14_centers, stats14_forwards, stats14_guards = filter_and_byPosition(result14)
stats15_centers, stats15_forwards, stats15_guards = filter_and_byPosition(result15)
stats16_centers, stats16_forwards, stats16_guards = filter_and_byPosition(result16)
stats17_centers, stats17_forwards, stats17_guards = filter_and_byPosition(result17)
stats18_centers, stats18_forwards, stats18_guards = filter_and_byPosition(result18)
stats21_centers, stats21_forwards, stats21_guards = filter_and_byPosition(result21)
"""

stats13_centers, stats13_forwards, stats13_guards = filter_and_byPosition(alldata13)
stats14_centers, stats14_forwards, stats14_guards = filter_and_byPosition(alldata14)
stats15_centers, stats15_forwards, stats15_guards = filter_and_byPosition(alldata15)
stats16_centers, stats16_forwards, stats16_guards = filter_and_byPosition(alldata16)
stats17_centers, stats17_forwards, stats17_guards = filter_and_byPosition(alldata17)
stats18_centers, stats18_forwards, stats18_guards = filter_and_byPosition(alldata18)
stats21_centers, stats21_forwards, stats21_guards = filter_and_byPosition(alldata21)

#print(stats21_forwards)
```
